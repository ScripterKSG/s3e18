```python
train = pd.read_csv('train.csv')
final = pd.read_csv('test.csv')
sample = pd.read_csv('sample_submission.csv')
```

```python
from sklearn import preprocessing
# from sklearn.model_selection import train_test_split

# split and make our training and testing sets for kfolds
y = train[['EC1', 'EC2']]
X = train.drop(['id','EC1','EC2','EC3','EC4','EC5','EC6'], axis=1)

scaler = preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X.values)
X = pd.DataFrame(X_scaled, columns=X.columns)

X.head()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

def get_model(n_input, n_output):
    model = keras.Sequential([
        layers.Dense(32, kernel_initializer='he_uniform', activation='relu', input_shape= [n_input]),
        layers.Dense(n_output, activation='sigmoid'),
    ])

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['binary_accuracy'],
    )
    
    return model
```

```python
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import accuracy_score

# evaluate a model using repeated k-fold cross-validation
results = list()
n_input, n_output = X.shape[1], y.shape[1]

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# enumerate folds
for train_ix, test_ix in cv.split(X):
    # prepare data
    X_train, X_test = X.iloc[train_ix], X.iloc[test_ix]
    y_train, y_test = y.iloc[train_ix], y.iloc[test_ix]
    # define model
    model = get_model(n_input, n_output)
    # fit model
    model.fit(X_train, y_train, verbose=0, batch_size=512, epochs=100)
    # make a prediction on the test set
    yhat = model.predict(X_test)
    # round probabilities to class labels
    yhat = yhat.round()
    acc = accuracy_score(y_test, yhat)
    # store result
    print('>%.3f' % acc)
    results.append(acc)
```

```python
# remove and save ids in final submission data
ids = final['id']
final = final.drop(['id'], axis=1)

scaler = preprocessing.StandardScaler()
final_scaled = scaler.fit_transform(final.values)
final = pd.DataFrame(final_scaled, columns=final.columns)

final.head()
```

```python
# do final prediction to generate submission
predictions = (model.predict(final) > 0.5).astype("int32")
predictions
```

```python
submit = pd.DataFrame(predictions, columns = ['EC1','EC2'])

submit['id'] = ids
submit = submit[['id','EC1','EC2']]

submit.head()
```
