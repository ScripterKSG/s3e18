```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample = pd.read_csv('sample_submission.csv')
```

```python
from sklearn import preprocessing

y_train = train[['EC1', 'EC2']]
X_train = train.drop(['id','EC1','EC2','EC3','EC4','EC5','EC6'], axis=1)

ids = test['id']
test = test.drop(['id'], axis=1)

#normalize training data and testing data
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X_train.values)
X_train = pd.DataFrame(X_scaled, columns=X_train.columns)

X_scaled_test = min_max_scaler.fit_transform(test.values)
test = pd.DataFrame(X_scaled_test, columns=test.columns)
```

```python
y_train.head()
```

```python
X_train.head()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

LogReg_pipeline = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),
            ])

# Training logistic regression model on train data EC1
LogReg_pipeline.fit(X_train, y_train['EC1'])

# Predicting
prediction_1 = LogReg_pipeline.predict(test)

# Training logistic regression model on train data EC 2
LogReg_pipeline.fit(X_train, y_train['EC2'])

# Predicting
prediction_2 = LogReg_pipeline.predict(test)
```

```python
submit = pd.DataFrame()
submit['id'] = ids
submit['EC1'] = prediction_1.tolist()
submit['EC2'] = prediction_2.tolist()
```
